```python
from truesight.util import (
    load_video,
    load_background,
    load_mask,
    metric_CIEDE2000,
    evaluate,
    load_triplets,
)    
import os
import cv2
import matplotlib.pyplot as plt
triplets = load_triplets("data/public/backgrounds", "data/public/videos","data/public/masks")

destination = "videos"

if not os.path.exists(destination):
    os.makedirs(destination)



for sample in triplets:
    video = load_video(sample[0])
    background = load_background(sample[2])
    mask = load_mask(sample[1])


    file_folder = os.path.join(destination, str(sample[0]).split("/")[-1].split(".")[0])
    print(file_folder)
    # break
    # save the vide frame by frame in the folder "videos" create folder for each video, name the images by there index
    if not os.path.exists(file_folder):
        os.makedirs(file_folder)
    
    for i, frame in enumerate(video):
        # save the frame as "videos/video_name/frame_index.png"
        # save numpy array as .png 
        plt.imsave(os.path.join(file_folder, f"fr_{i:04d}.jpeg"), frame[:,:,::-1], format="jpeg")
```
